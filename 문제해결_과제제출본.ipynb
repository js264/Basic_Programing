# 문제 7
- ### 비밀번호 강도 검사
- 길이가 8자 이상
- 영문자, 숫자가 포함되어야 함
- 특수문자 (!, @, $, #) 중 하나 이상 포함
- 모든 조건을 만족하면 '사용 가능' 출력

```python
pw = input('비밀번호 입력 : ')
has_num = False
has_special = False
has_english = False
not_has_korean = True

for ch in pw:
  # 숫자 검사
  if ch.isdigit():
    has_num = True
  # 특수문자 검사
  if ch in '!@#$':
    has_special = True
  # 영문자 검사
  if ch.isalpha() and ch.isascii():
    has_english = True
  # 한글 검사
  if '가' <= ch <= '힣': # ch에 한글이 포함되는지 (가 ~ 힣까지 포함되는지) 검사하고, 존재하면 flag를 False로 바꿈
    not_has_korean = False

# 조건 검사
if len(pw) >= 8 and has_num and has_special and has_english and not_has_korean:
  print('사용 가능')
else:
  print('사용 불가')
```

# 문제 8
- ### 학생 성적 순위 계산
- 학생들의 점수를 리스트로 저장한 후, 각 학생의 순위를 출력

```python
# 딕셔너리 활용
scores = {
    '민수':90,
    '지영':70,
    '철수':100,
    '영희':85
}

for k,v in scores.items():
  rank = 1            # 항상 rank를 1로 초기화시켜줌
  for i in scores:    # 현재 학생의 등수를 추출
    if scores[i] > v:
      rank += 1       # 만약 현재 key의 학생보다 높은 점수가 발견되면 rank를 1 증가 시킴
  print(k,":",v, '점 , ', rank, '등')
```

# 문제 9
- ### 학생별 최고 점수 찾기
- 학생 이름과 시험 점수가 튜플 형태로 저장되어 있다. 각 학생의 최고 점수를 출력

```python
# 딕셔너리 형태
scores = {
    '민수' : [80, 95],
    '지영' : [90, 88],
    '철수' : [70]
}

result = {}

for k,v in scores.items(): # scores.items()로 scores의 key값과 value값을 받아서 k, v로 저장
    result[k] = max(v) # key에 해당하는 value값의 최대값만 result에 저장

print(result)
```

# 문제 10
- ### 은행 계좌 관리
- 은행 계좌의 입금 금액(balance)과 출금 금액(money)를 입력 받는다.
- 단, 잔액이 부족하면 출금할 수 없다. 또한 출금 후 잔액이 1000원 미만이면 '최소 잔액 부족'이라는 메시지를 출력.

```python
mon = 0 # 초기 계좌 잔액 = 0

while True: # 무한 반복으로 계좌 관리를 연다.
    print('===== 은행 계좌 관리 =====\n1. 입금\n2. 출금\n3. 잔액 조회\n4. 종료')
    select = int(input("메뉴 선택 :")) # select로 계좌 관리 메뉴 번호를 선택하게 함

    # 입금
    if select == 1:
        balance = int(input("입금 금액 :")) # balance에 입금 금액을 받아서 mon에 저장해 계속 입금할 수 있게 만듦
        mon += balance
        print('입금 완료')
        print(f'현재 잔액 : {mon}\n')

    # 출금
    if select == 2:
        money = int(input("출금 금액 :"))
        if mon > money:
            if mon-money<1000: # 계좌 잔액이 1000 이하가 되는 경우
                print("최소 잔액 부족\n")
            else:
                mon -= money
                print('출금 완료')
                print(f'남은 잔액 : {mon}\n')
        else: # 계좌 잔액보다 출금 금액이 큰 경우
            print('잔액 부족\n')

    # 현재 잔액 조회
    if select == 3:
        print(f"현재 잔액 : {mon}\n")

    # 프로그램 종료
    if select == 4:
        print("프로그램 종료\n")
        break # break로 무한 반복문을 빠져나감
```
