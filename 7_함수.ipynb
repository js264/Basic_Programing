# 예제 : BMI

```python
# BMI 계산
def BMI(height, weight):
  bmi = weight/(height*height)
  return bmi

# BMI 평가
def result_BMI(bmi):
  if bmi<18.5:
    return '저체중'
  elif bmi<23:
    return '정상'
  elif bmi<25:
    return '과체중'
  elif bmi<30:
    return '경도비만'
  else:
    return '고도비만'


h = float(input('키 입력:'))/100
w = float(input('몸무게 입력:'))
bmi = BMI(h,w)

print(f'당신의 bmi 수치는 {bmi:.2f}입니다.')
print(f'당신은 {result_BMI(bmi)}입니다.')
```

# 예제 1

```python
# 소수 구하기

def primes(k):
  for i in range(2, k):
    if k % i == 0:
      return False
  return True

n = int(input("2~n까지 소수를 구할 수 입력: "))
print(f'2~{n}까지의 소수')
for k in range(2, n+1):
  if primes(k) == True:
    print(k, end=' ')
```

```python
# 에라토스테네스의 체

n = int(input('2부터 소수를 구할 수를 입력: '))
prime = [True] * (n+1) # 처음은 모두 소수라고 가정(True)
prime[0] = prime[1] = False # 0,1은 소수가 아님

# 큰 약수는 작은 약수와 항상 짝을 이루므로 루트n까지만 검사
for i in range(2, int(n**0.5) + 1):
  if prime[i]:
    for j in range(i*i, n+1, i):
      prime[j] = False

print(f'\n2~{n}까지의 소수')
print('-'*20)
for i in range(2, n+1):
  if prime[i]:
    print(i, end= ' ')
```

```python
# 함수로 만듦

def isprime(n):
  prime = [True] * (n+1) # 처음은 모두 소수라고 가정(True)
  prime[0] = prime[1] = False # 0,1은 소수가 아님

  # 큰 약수는 작은 약수와 항상 짝을 이루므로 루트n까지만 검사
  for i in range(2, int(n**0.5) + 1):
    if prime[i]: # 현재 수가 소수(True)이면
      for j in range(i*i, n+1, i):
        prime[j] = False # i의 배수 제거

  result = []
  for i in range(2, n+1):
    if prime[i]: # 소수만 저장
      result.append(i)
  return result

n = int(input('2부터 소수를 구할 수를 입력: '))
print('소수 : ', isprime(n))
```

# 예제 2

```python
def whatTriangle(a,b,c):
  if(a+b <= c or c+b <= a or a+c <= b):
    return '삼각형이 아님'
  elif(a==b and b==c):
    return '정삼각형'
  elif(a==b or b==c or a==c):
    return '이등변삼각형'
  elif(a**2+b**2==c**2 or a**2+c**2==b**2 or c**2+b**2==a**2):
    return '직각 삼각형'
  else:
    return '일반 삼각형'


a,b,c = map(int, input('세 변의 길이 입력: ').split())
print('%s\n'%whatTriangle(a,b,c))
```

# Mission 02 : 한영사전 구현

```python
dic1 = {"사랑" : "love",
        '사과' : 'apple',
        '하늘' : 'sky',
        '아기' : 'baby'}

def searchWord():
  han = input("검색할 한글 입력: ")
  if han in dic1:
    print(f'{han} : {dic1.get(han)}')
  else:
    print('없는 단어입니다.')

def addWord():
  han = input('추가할 한글 단어: ')
  eng = input("영어 단어 : ")
  dic1[han] = eng
  print('단어가 추가되었습니다.')
  print(dic1)

def deleteWord():
  print("삭제 전 데이터 :", dic1)
  han = input('삭제할 한글 단어: ')
  if han in dic1:
    print('삭제된 단어: ', dic1.pop(han))
  else:
    print('해당 단어가 없습니다.')
  print('삭제 후 데이터 :', dic1)

def printAll():
  print('===== 전체 사전 출력 =====')
  for k, v in dic1.items():
    print(k, ':', v)




def menu():
  while True:
    print('\n===== 한영 사전 메뉴 =====')
    print("1. 단어 검색")
    print("2. 단어 추가")
    print("3. 단어 삭제")
    print("4. 전체 출력")
    print("5. 종료\n")

    num = int(input('메뉴 선택 : '))
    if num == 1:
      searchWord()
    elif num ==2:
      addWord()
    elif num ==3:
      deleteWord()
    elif num ==4:
      printAll()
    elif num ==5:
      print('프로그램 종료')
      break
    else:
      print('잘못 입력했습니다.')

menu()
```

# Mission 03 : 재고 관리 구현

```python
store = {
    '사과' : 10,
    '바나나' : 5,
    '우유' : 7,
    '과자' : 15,
    '라면' : 20
}

# 메뉴 출력 함수
def display_menu1():
  print('\n===== 재고 관리 프로그램 =====')
  print('1. 물품 입고')
  print('2. 재고 확인')
  print('3. 물품 판매')
  print('4. 전체 재고 출력')
  print('5. 종료')
  return int(input('메뉴 선택: '))

# 물품 입고 함수
def add_stock1():
  print("\n[물품 입고]")
  print("종료하려면 z 입력\n")
  while True:
    item = input('입고 물품: ')
    if item == 'z':
      break
    count = int(input('입고량:'))
    store[item] = store.get(item,0) + count
    print(f'{item} {count}개 입고 완료')
    print(f'현재 재고: {store[item]}개\n')

# 재고 확인 함수
def check_stock1():
  print('\n[재고 확인]')
  print('종료하려면 엔터 입력\n')
  while True:
    item = input('찾는 물품: ')
    if item == '':
      break;
    if item in store:
      print(f'{item} 재고 : {store[item]}개')
    else:
      print('없는 물품입니다.')

# 물품 판매 함수
def sell_stock1():
  print('\n[물품 판매]')
  print('종료하려면 z 입력\n')
  while True:
    item = input('판매 물품: ')
    if item == 'z':
      break
    if item in store:
      sell = int(input('판매 수량: '))
      if store[item] >= sell:
        store[item] -= sell
        print("판매 완료!")
        print(f'현재 재고 : {store[item]}개')
      else:
        print('재고 부족')
    else:
      print('없는 물품입니다.')

# 전체 재고 출력 함수
def print_all1():
  print('\n ===== 전체 재고 현황 =====')
  if len(store) == 0:
    print('저장된 재고가 없습니다.')

  print('-'*30)
  print('물품명\t\t재고수량')
  print('-'*30)

  for key in sorted(store):
    print(f'{key}\t\t{store[key]}개')
  print('-'*30)

# 메인 함수
def main():
  while True:
    menu = display_menu1()
    if menu == 1:
      add_stock1()
    elif menu == 2:
      check_stock1()
    elif menu == 3:
      sell_stock1()
    elif menu == 4:
      print_all1()
    elif menu == 5:
      print('프로그램 종료')
      break
    else:
      print('잘못 입력했습니다.')


print_all1()
main()
```
