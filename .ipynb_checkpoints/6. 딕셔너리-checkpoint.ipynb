## 편의점 재고 관리하기

```python
store = {}
print("*** 물품별 재고량 입력 ***")
while True :
  item = input("입력 물품 ==> ")
  if item == "z":
    break
  count = int(input("재고량 ==>"))
  store[item] = count

#print(store)

#for k in store:
#  print(k)

for k,v in store.items():
  print(k,v)

#for k,v in store:
#  print(k,v) # -> 불가능
```

```python
store = {}
print("*** 물품별 재고량 입력 ***")

# 입고
while True :
  item = input("입력 물품 ==> ")
  if item == "z":
    break
  count = int(input("재고량 ==>"))
  store[item] = store.get(item,0) + count # 누적처리

# 재고 확인
print("\n*** 물품의 재고량 확인 ***")
while True :
  item = input("찾을 물품 ==> ")
  if item == "":
    break
  if item in store:
    print(f'{store[item]}개 남았어요')
  else:
    print("없는 물품입니다.")

# 판매
print('\n***물품 판매***')
while True:
  item = input('판매할 물품:')
  if item == "":
    break
  if item in store:
    sell = int(input('판매 수량:'))
    if store[item] >= sell:
      store[item] -= sell
      print(f'판매 완료! {store[item]}개 남았어요')
    else:
      print('재고가 부족합니다.')
  else:
    print('없는 물품입니다')
```

## Mission : 문자열, 딕셔너리 01

```python
str1 = "sky hello hi good morning"
str1 = str1.upper()
print(str1)
dic1={}

for ch in str1:
  if 'A' <= ch and ch <= 'Z':
    if ch in dic1:
      dic1[ch] += 1
    else:
      dic1[ch] = 1
        
for k in dic1.keys():
  print("%s --> %s"%(k, dic1[k]))

#for k,v in dic1.items:
#  print("%s --> %s"%(k, v))
```

# Mission : 문자열, 딕셔너리 02

```python
str1 = "sky hello hi good morning"
str1 = str1.upper()
print(str1)
dic1, list1 = {}, []

for ch in str1:
  if 'A' <= ch and ch <= 'Z':
    if ch in dic1:
      dic1[ch] += 1
    else:
      dic1[ch] = 1

list1 = sorted(dic1.items())
for char1, count1 in list1 :
  print("%s --> %s"%(char1, count1))

list2 = sorted(dic1)
print(list2)
```

# Mission : 문자열, 딕셔너리 03

```python
str1 = "sky hello hi good morning"
str1 = str1.upper()
print(str1)
dic1, list1 = {}, []

for ch in str1:
  if 'A' <= ch and ch <= 'Z':
    if ch in dic1:
      dic1[ch] += 1
    else:
      dic1[ch] = 1

list1 = list(dic1.items())
list1.sort(key = lambda x : x[1], reverse = True)
for char1, count1 in list1:
  print("%s --> %s" %(char1, count1))
```

# Mission : tuple, dictionary 01

```python
scores = (('이순신', 88, 95, 90), ('김유신', 85, 90, 95), ("홍길동", 70, 90, 80), ('허생원', 90, 90, 95))
student_dic = {}

for val in scores:
  avg = (val[1] + val[2] + val[3]) / 3
  student_dic[val[0]] = avg

print("이름        평균성적")
print("--------------------")
for k in student_dic.keys():
  print("{}   {:10.2f}".format(k, student_dic[k]))
```

# 음식 궁합 찾기

```python
foods = {"떡볶이" : "어묵",
         "짜장면" : '단무지',
         '라면' : '김치'}

print(foods.keys())
print(str(foods.keys()))
print(list(foods.keys()))
print("")
print(type(foods.keys()))
print(type(str(foods.keys())))
print(type(list(foods.keys())))
print("")
myfood = input(str(foods.keys()))
print(myfood,foods.get(myfood))
```

```python
foods = {"떡볶이" : "어묵",
         "짜장면" : '단무지',
         '라면' : '김치'}

while True:
  #myfood = input(foods + 중 좋아하는 음식은?)
  myfood = input(str(foods.keys()) + '중 좋아하는 음식은?')
  if myfood in foods :
    print("<%s> 궁합음식은 <%s>입니다."%(myfood, foods.get(myfood)))
  elif myfood == '끝':
    break
  else:
    print('그런 음식이 없습니다. 확인해보세요.')
```

# .update

```python
x = {'a':10, 'b':20, 'c':30, 'd':40}
print(x)
x.update(a = 90)
x.update(e = 50)
print(x)
x.update(a=900, f=60)
print(x)
```

```python
x = {'a':100, 'b':50, 'c': 60}
print(x)
x.update(a=50)
print(x)
x['a'] = 70
print(x)
x.update(a=11, b=44)
print(x)
```
