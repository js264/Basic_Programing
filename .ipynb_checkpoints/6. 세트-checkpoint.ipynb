```python
# 세트 pop : 임의의 요소 제거
num1 = {'one', 'two','three'}
print(num1)
a = num1.pop()
print(num1)
```

```python
# 차집합 (A-B)
num1 = {'one','two', 'three'}
num2 = {'four', 'five', 'three'}
num3 = num1-num2
print(num3)
```

```python
#difference(set1) : 두 집합 간의 차이가 포함된 집합을 반환
# 반환된 집합에는 두 집합이 아닌 첫 번째 집합에만 존재하는 항목이 포함
num3 = num1.difference(num2) # num3 = num1-num2 와 같은 의미
print(num3)
num4 = num2.difference(num1)
print(num4)
```

```python
# 교집합 (A&B)
num3 = num1 & num2
print(num3)
```

```python
# intersection(set1, set2, ... ) : 교집합
# 각 set에 모두 존재하는 항목을 포함하는 세트를 반환
num3 = num1.intersection(num2)
print(num3)
```

```python
# 합집합(A|B)
num3 = num1|num2
print(num3)
```

```python
# union(set1, set2, ...) : 합집합
# 두 세트의 모든 항목을 포함하는 세트를 반환합니다. 중복은 제외
num3 = num1.union(num2)
print(num3)
```

set 예제

```python
clubA = {'김유신', '박나나', '황희찬'}
clubB = {'박나나', '이순신', '최영'}

clubC = clubA.union(clubB)
print("동아리에 가입한 학생들의 모든 명단 : ", clubC)
print("A, B 동아리에 둘 다 가입한 학생의 명단 : ", clubA.intersection(clubB))
print("A 동아리 에서 B 동아리에 가입한 학생들의 명단을 제외 : ", clubA.difference(clubB))
print("B 동아리 에서 A 동아리에 가입한 학생들의 명단을 제외 : ", clubB.difference(clubA))

# A 동아리에 '홍길동' 회원이 새로 가입
clubA.add('홍길동')

# B 동아리에 '최영' 회원이 탈퇴
clubB.remove('최영')

# A, B 동아리 회원들을 각각 출력
print("최종 A 동아리 회원 : ", clubA)
print("최종 B 동아리 회원 : ", clubB)
```

```python
# 두 문자열의 공동적인 글자 찾기
'''
s1 = 'abcd'
print(set(s1))
'''

s1 = input('첫 번쨰 문자열 : ')
s2 = input('두 번쨰 문자열 : ')

list1 = set(s1)&set(s2)
#list2 = sorted(list1)
print('\n공통적인 글자 : ', end=" ")
for i in list1:
  print(i, end=" ")

s3 = set(s1).intersection(set(s2))
print(s3)
```

set 예제 : 입력된 문장 중에서 중복이 제거된 단어의 수를 구하세요. 단, 중복된 단어는 한 번만 표시. 세트를 이용

```python
t1 = input('문장 입력 : ')
w1 = t1.split()
print("분리된 단어의 개수 = ", len(w1))

u1 = set(w1) # 중복된 단어 제거
print("사용된 단어의 개수 = ", len(u1))
print("사용된 단어 = ", u1)

# MIssion : 입력된 문장 중에서 중복이 되지 않은 단어와 단어의 수를 구하세요
```
